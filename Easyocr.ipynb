```python
def show_results(results, width):
    for i, result in enumerate(results):
        # Сохраняем распознанный текст
        text = result[1].lower()
        proc = round(result[2] * 100, 2)
        # Выводим на экран распознанный текст
        print(f'\033[92m{i:>2}\033[0m | \033[94m{text:^{width}}\033[0m | \033[91m{proc}%\033[0m')
```

```python
import cv2
import matplotlib.pyplot as plt

# Path to your image
image_path = "/content/analiz.png"

# Load the image
image = cv2.imread(image_path)

# Convert the image to grayscale
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Display the grayscale image
plt.figure(figsize=(10, 10))
plt.imshow(gray_image, cmap='gray')
plt.axis('off')
plt.title("Grayscale Image")
plt.show()
```

```python
# Обнаружение текста
reader = easyocr.Reader(['ru'])
img = cv2.imread(test_img_path_ru)
results = reader.readtext(img)
```

```python
show_results(results, 30)
```
